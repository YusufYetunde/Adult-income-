```python
education_counts = df['education'].value_counts().reindex(['Bachelors', 'Masters']).fillna(0)
# visualize
plt.figure(figsize=(10, 5))
education_counts.plot(kind='bar')
plt.xlabel('Education Level')
plt.ylabel('Number of Persons')
plt.title('Number of Persons with Bachelors or Masters Degrees')
plt.show()
```

```python
#Replace Income values ['<=50K', '>50K'] with 0 and 1
df['income']= df['income'].replace({'<=50K':0, '>50K':1})
```

```python
sns.countplot(x='income', data=df)
```

```python
df['income'].replace(to_replace=['<=50K', '>50K'], value=[0, 1])
```

```python
# Which Workclass gets the highest salary?
df.groupby('workclass')['income'].mean().sort_values(ascending=False)
```

```python
workclass_income = df.groupby('workclass')['income'].mean().sort_values(ascending=False)

# Visualizing
plt.figure(figsize=(10, 5))
workclass_income.plot(kind='bar')
plt.xlabel('Workclass')
plt.ylabel('Average Income')
plt.title('Average Income by Workclass')
plt.xticks(rotation=45)
plt.show()
```

```python
# Who has better chance to get salary >50K, male or female?
df.groupby('gender')['income'].mean().sort_values(ascending=False)
```

```python
gender_income = df.groupby('gender')['income'].mean().sort_values(ascending=False)

# visualize
plt.figure(figsize=(10, 5))
gender_income.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Income')
plt.title('Average Income by Gender')
plt.xticks(rotation=0)
plt.show()
```
